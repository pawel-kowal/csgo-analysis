```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/user/Desktop/Data Analysis/archive/picks.csv')
df.head(5)
```

```python
t1_bans = df['t1_removed_1'].value_counts()
t1_bans
```

```python
t2_bans = df['t2_removed_1'].value_counts()
t2_bans
```

```python
bans = t1_bans.add(t2_bans, fill_value=0)
bans
```

```python
bans = bans.where(bans>1)
bans = bans.dropna()
print(f'Bans: {bans}')
```

```python
bans = bans.sort_values(ascending=False)
bans
```

```python
df_no_bo1 = df.loc[df['best_of'].isin(['3','5'])]
df_no_bo1
```

```python
t1_picks = df_no_bo1['t1_picked_1'].value_counts()
t2_picks = df_no_bo1['t2_picked_1'].value_counts()
picks = t1_picks.add(t2_picks)
picks= picks.dropna()
picks
```

```python
picks = picks.where(picks>54)
picks = picks.dropna()
picks
```

```python
picks = picks.sort_values(ascending=False)
picks
```

```python
bans_picks_corr = np.correlate(bans, picks)
bans_picks_corr
```

```python
### There is a quite significant correlaction between bans and picks, what is intuitional.
### Teams'd rather pick map that isn't banned too fast.
```
